1. Scrape the details of most viewed videos on YouTube from Wikipedia:
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
You need to find following details:

A) Rank

B) Name

C) Artist

D) Upload date

E) Views

```python
def most_viewed_videos_wiki(url):

    opts = Options()

    url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"

    opts.add_argument("user-agent=Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.66 Safari/537.36")

    driver.get(url)

    driver.find_element_by_xpath("//b").click()

    time.sleep(3)

    records = []
    for row in driver.find_elements_by_xpath("(//table[@class='wikitable sortable jquery-tablesorter'])[1]//tr")[1:-1]:
        temp = ''
        for element in row.find_elements_by_xpath(".//td"):        
            temp += element.text + '|'
        records.append(temp.split('|')[:-2])

    df = pd.DataFrame(records,columns=['rank','video_name','artist','date_uploaded','views(M)'])
    
    arranged_cols = ['rank','video_name','artist','date_uploaded','views(M)']
    
    df.video_name = df.video_name.apply(lambda x:x[:-4].strip('"'))
    df = df[arranged_cols]

    return df


most_viewed_videos_wiki("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

3. Scrape the details of selenium exception from guru99.com.¶
Url = https://www.guru99.com/

You need to find following details:

A) Name

B) Description

Note: - From guru99 home page you have to reach to selenium exception handling page through code

```python
def guru_sel_exc_handling(url):
    opts = Options()
    

    opts.add_argument("user-agent=Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.66 Safari/537.36")


    driver.get(url)

    driver.find_element_by_xpath("//a[@title='Selenium']").click()

    driver.find_element_by_xpath("//a[contains(@title,'Selenium Exception Handling')]").click()

    col_names,expn_l,des_l = [],[],[]

    for n,elements in enumerate(driver.find_elements_by_xpath("//table[@class='table table-striped']/tbody/*/td")):

        if n%2 == 0:
            if n == 0:
                col_names.append(elements.text)
            else:
                expn_l.append(elements.text)

        if n%2 == 1:
            if n == 1:
                col_names.append(elements.text)
            else:
                des_l.append(elements.text)

    df = pd.DataFrame(list(zip(expn_l,des_l)),columns=col_names) 

    return df

guru_sel_exc_handling("https://www.guru99.com/")
```

```python
# Opening Wikipedia webpage
url='https://archive.ics.uci.edu/'
driver.get(url)
time.sleep(1)
# Clicking on all datasets links
dataset=driver.find_element_by_xpath("//a[@href='datasets.php']")
dataset.click()
time.sleep(1)
# Creating empty lists
Dataset =[]
Data_Type =[]
Task =[]
Attribute_Type =[]
No_of_Instances =[]
No_of_Attribute =[]
Year =[]
# Scraping DataSet Name via Xpath
from tqdm import tqdm
try:
    dataset=driver.find_elements_by_xpath("//p[@class='normal']/b/a")
    for i in tqdm(dataset):
        Dataset.append(i.text)
        time.sleep(0.15)
except NoSuchElementException:
    Dataset.append('NA')
    
# Scraping Data Type via Xpath
try:
    Type=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]')
    for i in Type[1:]:
        Data_Type.append(i.text)
except NoSuchElementException:
    Data_Type.append('NA')
    
# Scraping Task via Xpath
try:
    task=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]')
    for i in task[1:]:
        Task.append(i.text)
except NoSuchElementException:
    Task.append('NA')
    
# Scraping Attribute_Type via Xpath
try:
    attribute_Type=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]')
    for i in attribute_Type[1:]:
        Attribute_Type.append(i.text)
except NoSuchElementException:
    Attribute_Type.append('NA')
    
# Scraping No_of_Instances via Xpath
try:
    no_of_Instances=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]')
    for i in no_of_Instances[1:]:
        No_of_Instances.append(i.text)
except NoSuchElementException:
    No_of_Instances.append('NA')

# Scraping No_of_Attribute via Xpath
try:
    no_of_Attribute=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]')
    for i in no_of_Attribute[1:]:
        No_of_Attribute.append(i.text)
except NoSuchElementException:
    No_of_Attribute.append('NA')
    
# Scraping Year via Xpath
try:
    year=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]')
    for i in year[1:]:
        Year.append(i.text)
except NoSuchElementException:
    Year.append('NA')
    
    


# Creating Data Frame for UCI Dataset
import numpy as np
UCI_Dataset=pd.DataFrame({'DataSet Title':pd.Series(Dataset),
                'Data Type':pd.Series(Data_Type),
                'Task':pd.Series(Task),
                'Attribute Type':pd.Series(Attribute_Type),
                'No of Instances':pd.Series(No_of_Instances),
                'No of Attribute':pd.Series(No_of_Attribute),
                'Year':pd.Series(Year)})
print('\033[1m'+' UCI Machine Learning Dataset:'+'\033[0m')
UCI_Dataset    
```

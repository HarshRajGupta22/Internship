#  Q 3 program to access the search bar and search button on images.google.com and scrap

# Q 5 a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google

```python
driver = webdriver.Chrome(r'H:/downloads/chromedriver_win32/chromedriver.exe')

#get the web page with given url
url = "https://www.google.com/maps/@16.4721906,74.435855,14z"
driver.get(url)

srch_box = driver.find_element_by_class_name("tactile-searchbox-input")  #fetching search bar

place = str(input("Enter place name, you want to search: "))  #Entering inputs to search for location

srch_box.send_keys(place)    

submit = driver.find_element_by_class_name("tactile-searchbox-input")
time.sleep(5)
submit.click()
current_url = driver.current_url   #coordinates are presents in URL so fetch current url

import re

lat = re.findall('(\d+(\.\d*)?)', current_url)[0]   #fetching latitude
long = re.findall('(\d+(\.\d*)?)', current_url)[1]     #fetching longitude

print("coordinates for the searched location,", place + " are: ")
print("Latitude :", lat[0])
print("Longitude :", long[0])
```

# Q 6 Write a program to scrap details of all the funding deals for second quarter (i.e Jan 21 – March 21) 


```python
#6
#importing webdriver
driver=webdriver.Chrome(r'H:/downloads/chromedriver_win32/chromedriver.exe')
time.sleep(4)
url="https://trak.in/india-startup-funding-investment-2015/"
driver.get(url)  #getting the funding page
time.sleep(4)

#creating empty boxex
date=[]
s_name=[]
industry=[]
sub=[]
location=[]
i_name=[]
i_type=[]
amount=[]
#getting the data of the month from july to september
for i in range(48,51):
    # Date
    dt = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[2]'.format(i))
    for d in dt:
        date.append(d.text)

    #startupname
    s=driver.find_elements_by_xpath('//*[@id="tablepress-{}"]/tbody/tr/td[3]'.format(i))
    for d in s:
        s_name.append(d.text)

    #industry
    a=driver.find_elements_by_xpath('//*[@id="tablepress-{}"]/tbody/tr/td[4]'.format(i))
    for d in a:
        industry.append(d.text)
    #sub
    su=driver.find_elements_by_xpath('//*[@id="tablepress-{}"]/tbody/tr/td[5]'.format(i))
    for d in su:
        sub.append(d.text)
    #location
    lo=driver.find_elements_by_xpath('//*[@id="tablepress-{}"]/tbody/tr/td[6]'.format(i))
    for d in lo:
        location.append(d.text)
    #i_name
    ina=driver.find_elements_by_xpath("//*[@id='tablepress-{}']/tbody/tr/td[7]".format(i))
    for d in ina:
        i_name.append(d.text)
    #i_type
    ty=driver.find_elements_by_xpath('//*[@id="tablepress-{}"]/tbody/tr/td[8]'.format(i))
    for d in ty:
        i_type.append(d.text)
    #amount
    am=driver.find_elements_by_xpath('//*[@id="tablepress-{}"]/tbody/tr/td[9]'.format(i))
    for d in am:
        amount.append(d.text)

time.sleep(2)
#creating dataframe
ans6=pd.DataFrame({})
ans6["Date"]=date
ans6["Startup_Name"]=s_name
ans6["Industry"]=industry
ans6["Sub"]=sub
ans6["Location"]=location
ans6["Investors name"]=i_name
ans6["Investment_type"]=i_type
ans6["Amount"]=amount
ans6
```

# Q 7 Write a program to scrap all the available details of best gaming laptops from digit.in.

```python
driver = webdriver.Chrome(r'H:/downloads/chromedriver_win32/chromedriver.exe')
#get the web page with given url
url = "https://www.digit.in/"
driver.get(url)
time.sleep(3)
#lets search for best laptop
best_gaming_lap = driver.find_element_by_xpath("//div[@class='listing_container']//ul//li[9]").click()
time.sleep(4)

#lets fetch info of laptops
lap_name = []
op_stm = []
display = []
processor = []
memory = []
weight = []
dimensions = []
graph_processor = []
prices = []
# lets fetch laptop names
name = driver.find_elements_by_xpath("//table[@id='summtable']//tr//td[1]")
for i in name:
    lap_name.append(i.text)
#fetching info related to operating system
try:
    os = driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[3]//td[3]")
    for i in os:
        op_stm.append(i.text)
except NoSuchElementException:
    pass
try:
    disp = driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[4]//td[3]")
    for i in disp:
        display.append(i.text)
except NoSuchElementException:
    pass
try:
    pro = driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[5]//td[3]")
    for i in pro:
        processor.append(i.text)
except NoSuchElementException:
    pass
try:
    mem = driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[6]//td[3]")
    for i in mem:
        memory.append(i.text)
except NoSuchElementException:
    pass
try:
    wgt = driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[7]//td[3]")
    for i in wgt:
        weight.append(i.text)
except NoSuchElementException:
    pass
try:
    dim = driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[8]//td[3]")
    for i in dim:
        dimensions.append(i.text)
except NoSuchElementException:
    pass
try:
    g_pro = driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[9]//td[3]")
    for i in g_pro:
        graph_processor.append(i.text)
except NoSuchElementException:
    pass
try:
    price = driver.find_elements_by_xpath("//td[@class='smprice']")
    for i in price:
        prices.append(i.text.replace('₹','Rs '))
except NoSuchElementException:
    pass

#creating dataframe
data = list(zip(lap_name,op_stm,display,processor,memory,weight,dimensions,graph_processor,prices))
df = pd.DataFrame(data, columns = ["Laptop Name","Operating System","Display","Processor","Memory","Weight(Kg)","Dimensions","Graphic-Processor","Price"])
df
```

# Q 9 program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video.

```python
driver = webdriver.Chrome(r'H:/downloads/chromedriver_win32/chromedriver.exe')
#get the web page with given url
url = "https://www.youtube.com/"
driver.get(url)
time.sleep(3)

#lets click on a video
clk_video = driver.find_element_by_xpath("/html/body/ytd-app/div/ytd-page-manager/ytd-browse/ytd-two-column-browse-results-renderer/div[1]/ytd-rich-grid-renderer/div[6]/ytd-rich-item-renderer[3]/div/ytd-rich-grid-media/div[1]/div/div[1]/h3/a").click()
time.sleep(8)

#Lets scroll the comments

for _ in range(1000):

    driver.execute_script("window.scrollBy(0,100)")

#fetching required data
comments = []
upvotes = []
times = []
try:
    comment = driver.find_elements_by_xpath("//*[@id='content-text']")[:500]
    for i in comment:
        comments.append(i.text)
except NoSuchElementException:
    comments.append('-')
time.sleep(3)

#fetching upvotes
try:
    up = driver.find_elements_by_xpath("//*[@id='vote-count-middle']")[:500]
    for i in up:
        upvotes.append(i.text)
except NoSuchElementException:
    comments.append('-')
    time.sleep(3)

#fetchin time when comment was posted
try:
    tim = driver.find_elements_by_xpath("//yt-formatted-string[@class='published-time-text above-comment style-scope ytd-comment-renderer']")[:500]
    for i in tim:
        times.append(i.text.replace('(edited)',''))
except NoSuchElementException:
    times.append('-')
time.sleep(3)

#creating dataframe
data = list(zip(comments,upvotes,times))
df = pd.DataFrame(data, columns = ["Comment","Upvotes","Published time"])
df
```
